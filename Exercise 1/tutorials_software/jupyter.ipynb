## Local computer environment
 
If you run Jupyter on your computer, it shares the R engine with the rest of the R applications, such as R studio(unless you have specifically created another R installation specifically for Jupyter - for example, two different Environments in Anaconda). The packages only need to be installed only once and then they will be available for all applications(what I install here is available in Rstudio)
 
**Use CTRL + ENTER to start the cell while staying in it.**
 
**Use SHIFT + ENTER to start the cell and jump to the next one below.**

## Display outputs
 
**You can split the current cell using CTRL + SHIFT + '-'**
 
**You can hide the input of the current cell with 'o'**
 
**The current output can be rolled up using SHIFT + 'o'**
 


```r
data = read.csv2(file="./data/aku.csv")
data
boxplot(data)
```

## Inserting and deleting new cells
 
**Use 'A' to insert a new cell before the current cell.**
 
**Use 'B' to insert a new cell after the current cell(also possible with a plus in the toolbar).**
 
**Use 'X' to delete the current cell(also using the scissors in the toolbar).**
 
You must have focus on the cells for these commands, but you must not have them open for editing - you can do so by clicking to the left of In []:
 


## Stop current calculation, restart kernel

**The current kernel can be restarted using 'Reset the kernel'(round arrow in the toolbar).**
